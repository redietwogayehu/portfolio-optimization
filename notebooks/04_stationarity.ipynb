```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

df = pd.read_csv("../data/processed/clean_stock_data.csv")

df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values(["Ticker", "Date"])

print(df.head())
print(df.columns)
```

```python
df = df.dropna(subset=["Daily_Return"])
```

ADF TEST FUNCTION

```python
def adf_test(series, name="Series"):
    result = adfuller(series.dropna())

    print(f"\n--- {name} ---")
    print(f"ADF Statistic: {result[0]:.4f}")
    print(f"p-value: {result[1]:.6f}")
    print("Critical Values:")
    for k, v in result[4].items():
        print(f"   {k}: {v:.4f}")

    if result[1] < 0.05:
        print("✅ Stationary (reject H0)")
    else:
        print("❌ Non-stationary (fail to reject H0)")
```

RUN STATIONARITY TEST

```python
tickers = df["Ticker"].unique()

for t in tickers:
    series = df[df["Ticker"] == t]["Daily_Return"]
    adf_test(series, f"{t} Returns")
```

PRICE vs RETURNS COMPARISON

```python
for t in tickers:
    price_series = df[df["Ticker"] == t]["Close"]
    adf_test(price_series, f"{t} Price")
```
